> Buatlah *table* seperti berikut dari data dinas sosial dan data *master* desa:

```python
# Import library yang dibutuhkan untuk mengakses contoh tabel
import pandas as pd

# Mengakses contoh tabel yang tersedia dalam format 'excel'
contoh_tabel = pd.read_excel(
    io='../docs/Data Scientist Techncial Test.xlsx', # Letak dan nama file dari contoh tabel
    sheet_name='bentuk table', # Nama sheet dari contoh tabel
    index_col='id' # Menggunakan kolom 'id' sebagai index dari contoh tabel
                   # Untuk menghilangkan redundansi dari index yang disediakan secara built-in oleh library Pandas
                   # Dengan 'id' yang berasal dari contoh tabel dengan fungsi yang serupa
)

# Menampilkan 'n' baris pertama dari contoh tabel untuk 'ease of viewing'
contoh_tabel.head(
    n=5
)
```

> *Note*: untuk *header* kolom menggunakan bentuk kecil semua dan isi data mengguanakan huruf besar semua.

Berikutnya, contoh tabel di atas akan dibandingkan dengan data dinas sosial dan data *master* desa yang tersedia pada folder `data`.

```python
# Membuat variabel instances dari data dinas sosial dalam format 'csv'
dinsos_csv = pd.read_csv(
    filepath_or_buffer='../data/Data Dinas Sosial.csv',
    sep=';', # Character yang digunakan sebagai delimiter/separator
    index_col='id'
)

# Dan dalam format 'excel'
dinsos_xlsx = pd.read_excel(
    io='../data/Data Dinas Sosial.xlsx',
    index_col='id'
)

# Membuat variabel instance dari data master desa
master_desa = pd.read_csv(
    filepath_or_buffer='../data/masterdesa.csv',
    sep=';'
)
```
