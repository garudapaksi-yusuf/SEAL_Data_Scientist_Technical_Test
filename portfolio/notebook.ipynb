## Section 3 | Data Cleansing
### Missing Value
* Sebelum membuat modifikasi lebih jauh pada `table` yang dibentuk, inspeksi dapat dilakukan untuk memeriksa keberadaan anomali dalam data.
* Selaras dengan instruksi berikutnya yang diberikan:

    > Lakukan *check* pada data tersebut, apakah ada kolom (dengan) nilai *null* atau tidak.

* Karena inspeksi ini akan dilakukan berulang kali, maka dibuat sebuah fungsi untuk mengurangi pengulangan kode Python yang sama berulang kali:

```python
def check_null():
    """Fungsi untuk mendeteksi dan menampilkan nilai null pada table

    Returns:
        DataFrame: DataFrame instance yang berisikan jumlah missing value pada tiap kolom table
    """
    df_null = pd.DataFrame(
        data=table\
            # Mendeteksi missing value (NA) pada tiap kolom dalam 'table'
            .isna()\
                # Kemudian menghitung jumlah missing value tersebut
                .sum(),
    # Mengganti nama kolom untuk memudahkan pemahaman
    ).rename(
        columns={0: 'NA (n)'}
    )
    return df_null
```

```python
# Deteksi pertama nilai null pada table
# Untuk mencari keberadaan anomali dalam data
check_null()
```

**Observasi:**
* Kolom dengan nilai *null* atau NA adalah kolom-kolom yang berasal dari `data_master_desa` (mulai dari `kode_provinsi_y` hingga `kemendagri_desa_kelurahan`).
* Tampilan berikut menunjukkan lokasi *missing value* tersebut dalam bentuk tabel:

```python
# Menunjukkan baris data dalam table dengan missing value
table[table.isna().any(axis=1)]
```

**Observasi**:
* Jumlah baris data `291` yang ditunjukkan pada tabel di atas sudah sesuai dengan yang dideteksi dari `check_null()`.
* Instruksi berikutnya memberikan arahan sebagai berikut:

    > Jika ada nilai *null*, lakukan *cleansing* data pada *table* yang sudah dilakukan *join* sehingga meminimalisir nilai *null*.

    > Buatlah *coding cleansing* dengan bahasa Python. Asumsikan akan melakukan *cleansing* data pada data yang jumlahnya besar dan beragam, sehingga buatlah algoritma *coding* yang bisa digunakan untuk banyak *case* data.
    
    > Buatlah penjelasan *cleansing* data yang Anda lakukan dari *problem* dan solusi yang diberikan.

* Sesuai petunjuk tersebut, beberapa kolom tersebut dapat diisi dengan menggunakan kolom serupa yang berasal dari `data_dinas_sosial` sebagai acuan —karena kolom-kolom ini terisi sempurna tanpa *missing value*—, sebagaimana ditunjukkan dalam bagian-bagian berikut:
    * Bagian pertama mendefinisikan fungsi yang diperlukan untuk mengganti nilai *null*, dan
    * Bagian kedua menerapkan fungsi tersebut pada pasangan kolom dengan nilai NA dan kolom tanpa nilai NA.

```python
def replace_null(value, replace):
    """Fungsi untuk mengganti nilai missing value

    Args:
        value (DataFrame): baris data dengan missing value
        replace (DataFrame): baris data tanpa missing value

    Returns:
        replace: jika value adalah missing value (None atau pd.isna())
        value: jika value selain missing value
    """    
    if value is None or pd.isna(value):
        return replace
    return value

def apply_replace_null(column_missing, column_filled):
    """Fungsi untuk menerapkan 'replace_null()' pada kolom table dengan missing value

    Args:
        column_missing (str): nama kolom yang berisikan missing value
        column_filled (str): nama kolom digunakan sebagai acuan untuk mengisi missing value tersebut
    """    
    table[column_missing] = table[[column_missing, column_filled]]\
        .apply(
            func=lambda row: replace_null(value=row[column_missing], replace=row[column_filled]),
            # Menggunakan 'axis' kedua atau kolom untuk menerapkan 'func' parameter
            # Sehingga 'func' diterapkan pada tiap baris dalam kolom tersebut
            axis=1
        )
```

```python
# Menjalankan 'apply_replace_null()' pada kolom table dengan missing value
apply_replace_null(column_missing='kode_provinsi_y', column_filled='kode_provinsi_x')
apply_replace_null(column_missing='nama_kabupaten_kota', column_filled='kabupaten')
apply_replace_null(column_missing='bps_nama_kecamatan', column_filled='kecamatan')
apply_replace_null(column_missing='kemendagri_nama_kecamatan', column_filled='kecamatan')
apply_replace_null(column_missing='bps_desa_kelurahan', column_filled='kelurahan')
apply_replace_null(column_missing='kemendagri_desa_kelurahan', column_filled='kelurahan')
```

```python
# Deteksi kedua nilai null pada table
# Setelah menjalankan 'apply_replace_null()' pada kolom tertentu dalam table
check_null()
```

**Observasi**:
* Fungsi `apply_replace_null()` berhasil dijalankan, dan *missing value* pada kolom yang ditetapkan telah terisi.

* Nama dari beberapa kolom seperti `bps_desa_kelurahan`, `kemendagri_desa_kelurahan`, dan `kategori` perlu disesuaikan dengan format nama kolom yang digunakan pada `contoh_tabel`.
* Demikian juga dengan urutan nama kolom pada `tabel` mengikuti urutan nama kolom `bentuk_table`

```python
table = table\
    .drop( # Menghapus kolom tertentu dari table
        columns='kode_provinsi_x'
    ).rename( # Mengganti nama kolom sesuai format pada `bentuk_table`
        columns={
            'bps_desa_kelurahan': 'bps_nama_desa_kelurahan',
            'kemendagri_desa_kelurahan': 'kemendagri_nama_desa_kelurahan',
            'kategori': 'kategori.1'
        }
    ).reindex( # Mengatur urutan dari table 
        labels=bentuk_table.columns, # Menggunakan daftar kolom dari 'bentuk_table'
        axis=1 # 'Axis' kedua atau kolom yang diurutkan sesuai parameter 'labels' sebelumnya
    )
```

**Observasi:**
Secara keseluruhan —mulai dari urutan dan nama kolom, isi, serta bentuk dari data— dapat dinyatakan bahwa `table` yang dibentuk telah menyerupai unsur-unsur yang dicontohkan dalam `bentuk_table`, kecuali pada poin-poin berikut:
1. Terkecuali isi dari kolom `kode_provinsi_y`, `kode_kabupaten_kota`, `bps_kode_kecamatan`, dan `bps_kode_desa_kelurahan` pada `table` yang memiliki sedikit perbedaan pada nilai dan format yang digunakan.
    * Perbedaan ini berasal dari keberadaan *missing value* `NaN` pada kolom-kolom ini sebelum dilakukan *cleansing*. [Karena `NaN` adalah bilangan bulat `float`](https://pandas.pydata.org/docs/dev/user_guide/integer_na.html), maka Pandas melakukan *enforcing* tipe data pada kolom tersebut yang seharusnya bertipe bilangan bulat *integer* menjadi *floating point*.
    * Karena `Missing Value` pada bagian sebelumnya berhasil dijalankan, maka langkah berikutnya cukup dengan mengubah tipe data `table` menjadi `integer`
2. Sedangkan `Nan` pada kolom `kategori` dikarenakan pengubahan nama kolom ini dari `kategori.1`.
    * Dan sebenarnya `kategori.1` tidak ditemukan pada `data_dinas_sosial` maupun `data_master_desa`. Sehingga penyesuaian format `table` mengikuti `bentuk_table` menciptakan kolom tambahan yang tidak berisikan data apapun.
    * Untuk menyesuaikan dengan instruksi `meminimalisir nilai null` sebelumnya, serta menjalankan prinsip [validitas data](https://en.wikipedia.org/wiki/Data_validation) agar struktur data yang dibuat tersusun secara logis dan konsisten, maka diputuskan untuk tidak menyertakan kolom ini pada `table`.

```python
table = table\
    .convert_dtypes( # Mengubah tipe data pada table
        convert_integer=True # Menjadi integer, jika dimungkinkan
    ).drop( # Menghapus kolom tertentu dari table
        columns='kategori'
    ).rename( # Mengubah nama kolom tertentu pada table
        columns={
            'kategori.1': 'kategori'
        }
    )
```

Dimensi dari *Table* tersebut dapat dibandingkan dengan sumber data dan contoh *table*:

```python
def view_shape(data):
    """Fungsi untuk melihat bentuk/dimensi dari data

    Args:
        data (DataFrame): input dalam format DataFrame

    Returns:
        DataFrame: output yang berisikan jumlah baris dan kolom dari input 
    """    
    df_shape = pd.DataFrame(
        data=[
            data.shape[0],
            data.shape[1]
        ],
        index=[
            'Baris',
            'Kolom'
        ],
        columns=[
            'Jumlah'
        ]
    )
    return df_shape
```

```python
# Menjalankan 'view_shape()' pada tiap data
print('Table baru:')
display(
    view_shape(
        data=table
    )
)

print('Data dinas sosial:')
display(
    view_shape(
        data=data_dinas_sosial
    )
)

print('Contoh table:')
display(
    view_shape(
        data=bentuk_table
    )
)
```

**Observasi:**
* Jumlah baris `11583` dari `table` yang telah dibentuk dan dibersihkan sudah sesuai dengan jumlah baris `11583` pada sumber data utama `data_dinas_sosial`.
* Jumlah kolom `24` dari `table` tersebut sudah sesuai dengan jumlah kolom `25` pada acuan `bentuk_table`, setelah menghilangkan 1 kolom `kategori.1`.
